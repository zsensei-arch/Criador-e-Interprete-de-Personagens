```python
import google.generativeai as genai

from google.colab import userdata
api_key = userdata.get ('secret_key')
genai.configure(api_key=api_key)

import random

# Configurando o modelo.
generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 0,
  "max_output_tokens": 108000,
}

safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_NONE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_NONE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_NONE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_NONE"
  },
]


class Personagem:
  def __init__(self, nome="", idade=0, personalidade="", biografia="", system_instruction=""):
    self.nome = nome
    self.idade = idade
    self.personalidade = personalidade
    self.biografia = biografia
    self.system_instruction = system_instruction

  def apresentar(self):
    print(f"Olá! Meu nome é {self.nome}.")
    if self.idade:
      print(f"Eu tenho {self.idade} anos.")
    if self.personalidade:
      print(f"Eu sou uma pessoa {self.personalidade}.")
    if self.biografia:
      print(f"Um pouco sobre mim: {self.biografia}")

def criar_personagem():
  nome = input("Qual o nome do personagem? ")
  idade = int(input("Qual a idade do personagem? "))
  personalidade = input("Descreva a personalidade do personagem: ")
  biografia = input("Conte um pouco sobre a história do personagem: ")
  system_instruction = f"Seu nome é {nome}. Sua idade é {idade} Anos de Idade. Sua personalidade é :{personalidade}. Sua biografia/história é :{biografia} . Você agirá e falará sómente como este personagem."
  return Personagem(nome, idade, personalidade, biografia, system_instruction)

def conversar(personagem):

  model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest",
                              generation_config=generation_config,
                              system_instruction=personagem.system_instruction,
                              safety_settings=safety_settings)
  chat = model.start_chat(history=[])
  print(f"\nAgora você pode conversar com {personagem.nome}. Para finalizar a conversa digite: sair")
  pergunta = input("Esperando pergunta: ")
  while pergunta != "sair":
    resposta = chat.send_message(pergunta)
    print("Resposta:", resposta.text, "\n\n")
    pergunta = input("Esperando pergunta: ")

# Loop principal do chatbot
while True:
  print("\nBem-vindo ao criador de personagens!")
  print("1. Criar novo personagem")
  print("2. Sair")

  escolha = input("Escolha uma opção: ")

  if escolha == "1":
    personagem = criar_personagem()
    personagem.apresentar()
    conversar(personagem)
  elif escolha == "2":
    break
  else:
    print("Opção inválida. Por favor, escolha uma opção válida.")
```
